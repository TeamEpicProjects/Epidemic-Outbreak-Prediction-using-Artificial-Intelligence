```python
# import libraries
from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# load dataset
df = pd.read_csv('../input/zikavirusdataset/classification_dataset.csv')
df.head()
```

```python
# Rows and cols
df.shape
```

```python
# toal duplicate values
df.duplicated().sum()
```

```python
# Remove duplicate rows
df = df.drop_duplicates(keep='first', inplace=False)
```

```python
df.shape
```

## Train Test Split

```python
train_data,test_data = train_test_split(df,test_size=0.40,random_state=0,stratify=df['Target'] ) #train test split in 60:40 ratio
```

```python
train_data.shape, test_data.shape
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
train_data['Target'].value_counts(), test_data['Target'].value_counts()
```

# Build Model

```python
label = 'Target'
save_path = 'agModels-predictClass'  # specifies folder to store trained models
predictor = TabularPredictor(label=label, path=save_path).fit(train_data)
```

```python
results = predictor.fit_summary(show_plot=True)
```

```python
predictor.leaderboard(train_data, silent=True)
```

# Make prediction

```python
y_test = test_data[label]  # values to predict
test_data_nolab = test_data.drop(columns=[label])  # delete label column to prove we're not cheating
test_data_nolab.head()
```

```python
predictor = TabularPredictor.load(save_path)  # unnecessary, just demonstrates how to load previously-trained predictor from file

y_pred = predictor.predict(test_data_nolab)
print("Predictions:  \n", y_pred)
```

```python
perf = predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True)
perf
```

```python
predictor.leaderboard(test_data, silent=True)
```

```python
pred_probs = predictor.predict_proba(test_data_nolab)
pred_probs.head(5)
```

```python
from sklearn.metrics import accuracy_score
y_pred_light = predictor.predict(test_data)
score = accuracy_score(y_test,y_pred_light)
print(score)
```

```python
# Accuracy with LightGBM
from sklearn.metrics import accuracy_score
y_pred_light = predictor.predict(test_data, model='LightGBM')
score = accuracy_score(y_test,y_pred_light)
print(score)
```

```python
# Accuracy with WeightedEnsemble_L2
from sklearn.metrics import accuracy_score
y_pred_light = predictor.predict(test_data, model='WeightedEnsemble_L2')
score = accuracy_score(y_test,y_pred_light)
print(score)
```

WeightedEnsemble_L2 is the best model with 60.71 accuracy
