# Adding weather data based on 7 days incubation period

```python
import pandas as pd
from glob import glob
import datetime
pd.pandas.set_option('display.max_columns',None)
```

```python
stock_files = sorted(glob('Weather_dataset/incubation/*.csv'))
data = pd.concat((pd.read_csv(file, encoding='cp1252')
                      for file in stock_files), ignore_index=True)
data.to_csv("Weather_dataset.csv",index=False)
print("Success...")
```

```python
weather = pd.read_csv('Weather_dataset.csv')
weather.head()
```

```python
weather = weather.drop(['moonrise', 'moonset','sunrise','sunset'], axis=1 )
```

```python
weather = weather.rename(columns = {'date_time':'incubation_date'})
```

```python
weather.head()
```

## Adding to final dataset

```python
df = pd.read_csv('final_dataset.csv')
df.head()
```

```python
df['report_date'] = pd.to_datetime(df['report_date'], format="%Y-%m-%d")
df['incubation_date'] = df['report_date'] -  pd.to_timedelta(7, unit='d')
```

```python
df.head()
```

```python
weather.columns
```

```python
df.columns
```

```python
df = df.drop(['maxtempC', 'mintempC', 'totalSnow_cm', 'sunHour',
       'uvIndex', 'moon_illumination', 'DewPointC', 'FeelsLikeC', 'HeatIndexC',
       'WindChillC', 'WindGustKmph', 'cloudcover', 'humidity', 'precipMM',
       'pressure', 'tempC', 'visibility', 'winddirDegree', 'windspeedKmph'], axis=1 )
```

```python
df['incubation_date'] = df['incubation_date'].apply(lambda x: datetime.date.strftime(x, '%Y-%m-%d'))
```

```python
df.head()
```

```python
# location wise population density_per_km 
final_df = pd.merge(df, weather, on=['incubation_date','location'], how='left') 
final_df.head()
```
