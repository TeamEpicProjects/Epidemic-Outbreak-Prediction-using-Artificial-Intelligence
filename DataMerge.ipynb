# Merge all dataset county wise
Zika viruse > Longitude & Altitude > Population Density > Weather

```python
# import library
import pandas as pd
import numpy as np
from glob import glob
import re
import datetime

import warnings
warnings.simplefilter("ignore")

pd.pandas.set_option('display.max_columns',None)
```

```python
zika = pd.read_csv('Zika_dataset/Clean_data/*.csv')
zika.head()
```

```python
stock_files = sorted(glob('Zika_dataset/Clean_data/*.csv'))
data = pd.concat((pd.read_csv(file, encoding='cp1252')
                      for file in stock_files), ignore_index=True)
data.to_csv("Zikavirus_dataset.csv",index=False)
print("Success...")
```

```python
df = pd.read_csv('Weather_dataset/Colombia_weather.csv')
df.head()
```

```python
df.shape
```

```python
# location wise population density_per_km 
new_df = pd.merge(zika, df, on=['date_time','location'], how='left') 
new_df.head()
```

```python
dense = pd.read_csv('06_population_density.csv')
dense.head()
```

```python
# location wise population density_per_km 
final_df = pd.merge(new_df, dense, on=['location'], how='left') 
final_df.head()
```

```python
latlog = pd.read_csv('01_latitude_longitude_google.csv')
latlog.head()
```

```python
# location wise population density_per_km 
final_dataset = pd.merge(final_df, latlog, on=['location'], how='left') 
final_dataset.head()
```

```python
final_dataset.to_csv('Final_data/Colombia.csv', index=False )
```
