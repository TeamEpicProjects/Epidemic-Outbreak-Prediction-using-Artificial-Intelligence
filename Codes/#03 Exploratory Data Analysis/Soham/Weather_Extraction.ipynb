```python
# import library
import pandas as pd
import numpy as np

import re
import datetime

import warnings
warnings.simplefilter("ignore")

pd.pandas.set_option('display.max_columns',None)
```

```python
zika = pd.read_csv('infection_ecuador.csv')
latlong = pd.read_csv('latlong.csv', usecols=['location', 'latitude', 'longitude'])
popden = pd.read_csv('popden.csv')
```

```python
latlong
```

```python
arg_latlong = latlong[latlong["location"].str.startswith("Ecuador")]
arg_latlong
```

```python
arg_popden = popden[popden["location"].str.startswith("Ecuador")]
arg_popden
```

```python
df_arg1 = pd.merge(zika, arg_popden, on=['location'])
df_arg1
```

```python
df_arg = pd.merge(df_arg1, arg_latlong, on=['location'])
df_arg
```

```python
# df_arg.to_csv('argentine_population_latlong.csv')
```

```python
arg_weather = pd.read_csv('ecuador_weather.csv')
arg_weather
```

```python
new = arg_weather.rename(columns={'date_time': 'report_date'})
new.head()
```

```python
df_argentina = pd.merge(df_arg, new, on=['location', 'report_date'])
df_argentina
```

```python
df_argentina.to_csv('merged_ecuador.csv', index=False)
```
