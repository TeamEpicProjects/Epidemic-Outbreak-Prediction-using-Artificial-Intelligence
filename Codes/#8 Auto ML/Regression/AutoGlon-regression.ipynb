```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('regression_dataset.csv')
data.head()
```

```python
data.describe()
```

```python
data = data[['maxtempC', 'mintempC', 'windspeedKmph', 'year', 'month', 'density_per_km', 'precipMM', 'cloudcover',
       'humidity','DewPointC', 'latitude', 'longitude','FeelsLikeC', 'winddirDegree', 'sunHour','cases']]
data.head()
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data, test_size=0.35, random_state=10)
```

```python
train_data.shape, test_data.shape
```

```python
target_column = 'cases'
print("Summary of target variable: \n", train_data[target_column].describe())
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
predictor = TabularPredictor(label=target_column, path="regModels-predictCase").fit(train_data, presets='best_quality',ag_args_fit={'num_gpus': 1})
```

```python
performance = predictor.evaluate(test_data)
```

```python
performance
```

```python
predictor.leaderboard(test_data, silent=True)
```

```python
y_test = test_data[target_column]  # values to predict
test_data_nolab = test_data.drop(columns=[target_column])  # delete label column to prove we're not cheating
test_data_nolab.head()
```

```python
predictor = TabularPredictor.load('regModels-predictCase')  # unnecessary, just demonstrates how to load previously-trained predictor from file

y_pred = predictor.predict(test_data_nolab)
print("Predictions:  \n", y_pred)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
ypred = predictor.predict(test_data)

RMSE = np.sqrt(mean_squared_error(y_test, ypred))
print('The value of RMSE is ...........', RMSE)

r_square = r2_score(y_test, ypred)
print('The value of Rsquare is ........', r_square)
```
