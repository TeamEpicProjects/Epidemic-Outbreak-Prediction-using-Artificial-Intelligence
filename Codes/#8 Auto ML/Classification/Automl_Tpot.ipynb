```python
from tpot import TPOTClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# load dataset
df = pd.read_csv('../input/zikavirusdataset/classification_dataset.csv')
df.head()
```

```python
# Rows and cols
df.shape
```

```python
# toal duplicate values
df.duplicated().sum()
```

```python
# Remove duplicate rows
df = df.drop_duplicates(keep='first', inplace=False)
```

```python
df.shape
```

# Train Test Split

```python
X = df.drop('Target',axis=1) #independent dimensions  
y = df['Target'] #selecting target column
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.35,random_state=1)
```

```python
X_train.shape, y_train.shape,X_test.shape, y_test.shape
```

# Build Model

```python
tpot = TPOTClassifier( random_state=10,verbosity=3,generations=5, population_size=50, n_jobs=-1,config_dict='TPOT NN')
tpot.fit(X_train, y_train)
```

```python
tpot.fitted_pipeline_
```

```python
tpot.score(X_test, y_test)
```

```python
tpot.export('tpot_digits_pipeline.py')
```

```python
from sklearn.metrics import accuracy_score
y_predi = tpot.predict(X_test)
score = accuracy_score(y_test,y_predi)
print(score)
```
