# Cleaning dataset one by one (Country wise)

```python
# import library
import pandas as pd
import numpy as np
from glob import glob
import re
import datetime

import warnings
warnings.simplefilter("ignore")

pd.pandas.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('Final_data/Colombia.csv')
df.head()
```

```python
df = df.drop(['moonrise', 'moonset','sunrise','sunset'], axis=1 )
```

```python
df = df.drop(['moonrise', 'moonset','sunrise','sunset','country','province','county','city','location_type'], axis=1 )
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = df.dropna(subset=['latitude'])
```

```python
df = df.dropna(subset=['winddirDegree'])
```

```python
df['cases'].fillna(df['cases'].median(), inplace=True)
```

```python
df.cases.median()
```

```python
len(df.columns)
```

```python
df.columns
```

```python
df = df.rename(columns = {'date_time':'report_date'})
```

```python

column_names = ['report_date', 'location', 'data_field', 'density_per_km',
       'latitude', 'longitude', 'maxtempC', 'mintempC', 'totalSnow_cm',
       'sunHour', 'uvIndex', 'moon_illumination', 'DewPointC', 'FeelsLikeC',
       'HeatIndexC', 'WindChillC', 'WindGustKmph', 'cloudcover', 'humidity',
       'precipMM', 'pressure', 'tempC', 'visibility', 'winddirDegree',
       'windspeedKmph','cases']

df = df.reindex(columns=column_names)
```

```python
df.head()
```

```python
df.to_csv('Final_data/Colombia.csv', index=False )
```

# Combine all countries zika dataset

```python
stock_files = sorted(glob('Final_dataset/*.csv'))
data = pd.concat((pd.read_csv(file, encoding='cp1252')
                      for file in stock_files), ignore_index=True)
data.to_csv("final_dataset.csv",index=False)
print("Success...")
```
